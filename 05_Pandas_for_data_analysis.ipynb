```python
import numpy as np
import pandas as pd
```

Pandas missing values

```python
d = {'A':[1,2,np.nan],'B':[2,np.nan,np.nan],'C':[7,8,9]}

df = pd.DataFrame(d)
df
```

```python
df.dropna()
```

```python
# Use inplace = True

df
```

```python
df.fillna(0)
```

```python
df
```

```python
df['A'].fillna(df['A'].mean())
```

Group By Method

```python
d = {'company':['Goog','Goog','Msft','Msft','FB','FB'],
     'person':['leo','theago','cristiano','sem','donny','luis'],
     'sales':[200,190,170,220,150,210]}
```

```python
df = pd.DataFrame(d)

df
```

```python
bycom = df.groupby('company')

bycom
```

```python
bycom.mean()
```

```python
bycom.sum()
```

```python
bycom.std()
```

```python
bycom.sum().loc['FB']
```

```python
df.groupby('company').sum().loc['FB']
```

```python
df.groupby('company').count()
```

```python
df.groupby('company').max()
```

```python
df.groupby('company').min()
```

```python
df.groupby('company').describe()
```

```python
df.groupby('company').describe().transpose()['FB']
```
