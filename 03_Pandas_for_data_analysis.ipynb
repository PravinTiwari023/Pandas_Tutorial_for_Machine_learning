```python
import numpy as np
import pandas as pd
```

Coditional selection

```python
df = pd.DataFrame(np.random.randn(5,4),['A','B','C','D','E'],['column 1','column 2','column 3','column 4'])

df
```

```python
df > 0
```

```python
booldf = df > 0

df[booldf]
```

```python
df['column 1'] > 0
```

```python
df[df['column 1'] > 0]
```

```python
new_df = df[df['column 1'] < 0]

new_df
```

```python
new_df['column 4']
```

```python
# we can do this in single command also

df[df['column 1']<0]['column 4']
```

```python
df[df['column 1']<0][['column 2','column 3']]
```

```python
# we directly can not use "and" and "or" it will show series is ambigious
# instead of this we can use "&" and "|"

df[(df['column 1']<0) & (df['column 3']<0)]
```
