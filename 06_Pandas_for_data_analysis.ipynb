```python
import numpy as np
import pandas as pd
```

Merging, Joining and Concatinating

```python
df1 = pd.DataFrame({'A':['A1','A2','A3','A4'],
                    'B':['B1','B2','B3','B4'],
                    'C':['C1','C2','C3','C4'],
                    'D':['D1','D2','D3','D4']},
                   index=[1,2,3,4])
```

```python
df2 = pd.DataFrame({'A':['A5','A6','A7','A8'],
                    'B':['B5','B6','B7','B8'],
                    'C':['C5','C6','C7','C8'],
                    'D':['D5','D6','D7','D8']},
                   index=[5,6,7,8])
```

```python
df3 = pd.DataFrame({'A':['A9','A10','A11','A12'],
                    'B':['B9','B10','B11','B12'],
                    'C':['C9','C10','C11','C12'],
                    'D':['D9','D10','D11','D12']},
                   index=[9,10,11,12])
```

```python
df1
```

```python
df2
```

```python
df3
```

### concatinate

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
left = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
```

```python
right = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

Merge

```python
pd.merge(left,right,how='inner',on='Key')
```

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index = ['K0','K1','K2'])
```

```python
right = pd.DataFrame({'C':['C0','C2','C3'],
                     'D':['D0','D2','D3']},
                    index = ['K0','K2','K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
